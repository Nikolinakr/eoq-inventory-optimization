## 1. Introduction

Inventory management is one of the main challenges in e-commerce. Companies must decide how much inventory to order and maintain in stock in order to satisfy customer demand while at the same time controlling the costs associated with storing these goods. Poor planning of inventory can lead both to product shortages and loss of customers, as well as to overstocking, increased storage costs, and deterioration of the company’s working capital.

The main issue in inventory management is finding a balance between:

• **Ordering cost** – the administrative and logistical costs that arise each time a new order is placed.

• **Holding cost** – the costs of storing inventory over a given period of time, including warehouse space, the risk of product obsolescence or damage, as well as the impact on the company’s working capital.

This requires the use of various mathematical models that support the optimization and management of inventory. One of the most commonly used models is **Economic Order Quantity (EOQ)**, which determines the optimal order quantity that minimizes the total cost of inventory management.

## 2. Problem Formulation

In formulating the **Economic Order Quantity (EOQ)** model, several main variables are used to describe the inventory management system.

Definition of the variables:

$D$ (annual demand) – The variable $D$ represents the total quantity of a given product that is demanded or sold within one year. It is measured in units per year. In the EOQ model, demand is assumed to be constant, known, and uniform. This means that seasonality of sales, changes in the market, or changes in consumer behavior are not taken into account. The parameter DDD is essential because the greater the demand, the more goods will need to be supplied, which directly affects the number of orders and the total costs.

$Q$ (order quantity) – The variable $Q$ represents the quantity of products ordered in a single order. If its value is small, this means that orders will be placed more frequently, which will increase ordering costs. If its value is high, the orders will be larger, which will increase holding costs. In the EOQ model, $Q$ is the variable that is subject to optimization. The objective is to find the value of $Q$ at which the sum of ordering costs and holding costs is minimized.

$S$ (cost per order) – The variable $S$ denotes the cost associated with placing a single order, regardless of its size. It is measured in monetary units, and for the purposes of the model, it is assumed to be fixed.

$H$ (holding cost) – The variable $H$ denotes the cost of holding one unit of product for a period of one year. It is important to note that, unlike $S$, where the number of products was not relevant, here the key factor is the quantity rather than the frequency of replenishment. The more goods there are in the warehouse, the higher the cost will be. In the model, the cost is assumed to be fixed for each unit.

The four variables $D$, $Q$, $S$, and $H$ form the basis of the **EOQ** model and determine the structure of total costs. Their combination allows the total cost function to be derived and the optimal quantity that minimizes these costs to be determined.
The objective is to determine the value of $Q$ that minimizes the total annual inventory-related cost.

### Cost Components

#### 1. Ordering Cost (per year)

If the annual demand is $D$ and each order contains $Q$ units, then the number of orders placed in one year is equal to the demand divided by the number of units in each order:
$$
\frac{D}{Q}
$$

This value indicates how many times per year orders must be placed in order to meet the required demand.

Since each order costs $S$, the total annual ordering costs are:

$$
\text{Ordering Cost}(Q) = \frac{D}{Q}S
$$
From this relationship, it can be seen that ordering costs decrease as the order quantity increases.

#### 2. Holding Cost (per year)

Under the classical assumptions of the **EOQ** model, the inventory level decreases linearly from $Q$ to $0$ between two consecutive deliveries. Therefore, the average inventory level is:
$$
\frac{Q}{2}
$$

If the holding cost per unit per year is $H$, then the total annual holding cost is:

$$
\text{Holding Cost}(Q) = \frac{Q}{2}H
$$
Unlike ordering costs, holding costs increase as the order quantity increases.

### Total Cost Function

Ordering costs and holding costs have opposite behavior with respect to the order quantity. It is precisely this balance and interaction that is analyzed in the **EOQ** model. Accordingly, total costs are equal to the sum of ordering costs and holding costs.

$$
TC(Q) = \frac{D}{Q}S + \frac{Q}{2}H
$$

Graphically, the total cost function has a **U-shaped form**. The minimum point of the U-shaped curve is where the ordering cost equals the holding cost. At this point lies the optimal quantity $Q$.

### Optimization Objective

Formally, the optimization objective can be expressed as follows:

$$
Q^* = \arg\min_{Q > 0} TC(Q)
$$

This expression means that we are looking for a positive value of $Q$ at which the total cost function $TC(Q)$ reaches its minimum.
