## 1. Introduction

Inventory management is one of the main challenges in e-commerce. Companies must decide how much inventory to order and maintain in stock in order to satisfy customer demand while at the same time controlling the costs associated with storing these goods. Poor planning of inventory can lead both to product shortages and loss of customers, as well as to overstocking, increased storage costs, and deterioration of the company’s working capital.

The main issue in inventory management is finding a balance between:

• **Ordering cost** – the administrative and logistical costs that arise each time a new order is placed.

• **Holding cost** – the costs of storing inventory over a given period of time, including warehouse space, the risk of product obsolescence or damage, as well as the impact on the company’s working capital.

This requires the use of various mathematical models that support the optimization and management of inventory. One of the most commonly used models is **Economic Order Quantity (EOQ)**, which determines the optimal order quantity that minimizes the total cost of inventory management.

## 2. Problem Formulation

In formulating the **Economic Order Quantity (EOQ)** model, several main variables are used to describe the inventory management system.

Definition of the variables:

$D$ (annual demand) – The variable $D$ represents the total quantity of a given product that is demanded or sold within one year. It is measured in units per year. In the EOQ model, demand is assumed to be constant, known, and uniform. This means that seasonality of sales, changes in the market, or changes in consumer behavior are not taken into account. The parameter DDD is essential because the greater the demand, the more goods will need to be supplied, which directly affects the number of orders and the total costs.

$Q$ (order quantity) – The variable $Q$ represents the quantity of products ordered in a single order. If its value is small, this means that orders will be placed more frequently, which will increase ordering costs. If its value is high, the orders will be larger, which will increase holding costs. In the EOQ model, $Q$ is the variable that is subject to optimization. The objective is to find the value of $Q$ at which the sum of ordering costs and holding costs is minimized.

$S$ (cost per order) – The variable $S$ denotes the cost associated with placing a single order, regardless of its size. It is measured in monetary units, and for the purposes of the model, it is assumed to be fixed.

$H$ (holding cost) – The variable $H$ denotes the cost of holding one unit of product for a period of one year. It is important to note that, unlike $S$, where the number of products was not relevant, here the key factor is the quantity rather than the frequency of replenishment. The more goods there are in the warehouse, the higher the cost will be. In the model, the cost is assumed to be fixed for each unit.

The four variables $D$, $Q$, $S$, and $H$ form the basis of the **EOQ** model and determine the structure of total costs. Their combination allows the total cost function to be derived and the optimal quantity that minimizes these costs to be determined.
The objective is to determine the value of $Q$ that minimizes the total annual inventory-related cost.

### Cost Components

#### 1. Ordering Cost (per year)

If the annual demand is $D$ and each order contains $Q$ units, then the number of orders placed in one year is equal to the demand divided by the number of units in each order:
$$
\frac{D}{Q}
$$

This value indicates how many times per year orders must be placed in order to meet the required demand.

Since each order costs $S$, the total annual ordering costs are:

$$
\text{Ordering Cost}(Q) = \frac{D}{Q}S
$$
From this relationship, it can be seen that ordering costs decrease as the order quantity increases.

#### 2. Holding Cost (per year)

Under the classical assumptions of the EOQ model, the inventory level decreases linearly from $Q$ to $0$ between two consecutive deliveries. Therefore, the average inventory level is:
$$
\frac{Q}{2}
$$

If the holding cost per unit per year is $H$, then the total annual holding cost is:

$$
\text{Holding Cost}(Q) = \frac{Q}{2}H
$$
Unlike ordering costs, holding costs increase as the order quantity increases.

### Total Cost Function

Ordering costs and holding costs have opposite behavior with respect to the order quantity. It is precisely this balance and interaction that is analyzed in the EOQ model. Accordingly, total costs are equal to the sum of ordering costs and holding costs.

$$
TC(Q) = \frac{D}{Q}S + \frac{Q}{2}H
$$

Graphically, the total cost function has a **U-shaped form**. The minimum point of the U-shaped curve is where the ordering cost equals the holding cost. At this point lies the optimal quantity $Q$.

### Optimization Objective

Formally, the optimization objective can be expressed as follows:

$$
Q^* = \arg\min_{Q > 0} TC(Q)
$$

This expression means that we are looking for a positive value of $Q$ at which the total cost function $TC(Q)$ reaches its minimum.

To find this value, methods from mathematical analysis are used, which will be examined and solved in the next section.

## 3. Mathematical Derivation of EOQ

After the total annual cost function has been formulated, the next step is to determine the value of the order quantity $Q$ at which these costs are minimized. This can be done by using methods from mathematical analysis, by examining the behavior of the total cost function.

The total annual cost function in the EOQ model has the following form:

$$
TC(Q) = \frac{D}{Q}S + \frac{Q}{2}H
$$

where:

- $D$ – annual demand for the product  
- $Q$ – order quantity  
- $S$ – cost per order  
- $H$ – holding cost per unit per year  

This function consists of two main components.  

The first term,

$$
\frac{D}{Q}S
$$

represents the annual ordering costs. These depend on the number of orders that must be placed during the year.

The second term,

$$
\frac{Q}{2}H
$$

represents the annual holding costs and depends on the average quantity of goods kept in inventory.

The objective is to find the value of $Q$ for which the sum of these two types of costs is minimized. Since the order quantity cannot be negative, we consider only values for which:

$$
Q > 0
$$

---

### First Derivative

To determine where the total cost function reaches its minimum, we compute the first derivative with respect to $Q$.

$$
\frac{dTC}{dQ} =
\frac{d}{dQ}\left(\frac{D}{Q}S + \frac{Q}{2}H\right)
$$

Since $D$, $S$, and $H$ are constants, the two terms can be differentiated separately.

Derivative of the first term:

$$
-\frac{DS}{Q^2}
$$

Derivative of the second term:

$$
\frac{H}{2}
$$

Therefore, the first derivative of the total cost function is:

$$
\frac{dTC}{dQ} =
-\frac{DS}{Q^2} + \frac{H}{2}
$$

---

### Finding the Optimal Point

To find the minimum of the function, the first derivative is set equal to zero:

$$
-\frac{DS}{Q^2} + \frac{H}{2} = 0
$$

Rearranging the equation:

$$
\frac{H}{2} = \frac{DS}{Q^2}
$$

Multiplying both sides by \(Q^2\):

$$
\frac{H}{2}Q^2 = DS
$$

Solving for \(Q^2\):

$$
Q^2 = \frac{2DS}{H}
$$

Taking the square root of both sides:

$$
Q = \sqrt{\frac{2DS}{H}}
$$

Since order quantity must be positive, we take the positive root:

$$
Q^* = \sqrt{\frac{2DS}{H}}
$$

---

### Economic Order Quantity

The resulting formula

$$
Q^* = \sqrt{\frac{2DS}{H}}
$$

is known as the **Economic Order Quantity (EOQ)**.

This formula shows how the optimal order quantity depends on the key parameters of the model:

- Higher **annual demand $D$** increases the optimal order quantity  
- Higher **ordering cost $S$** increases the optimal order quantity  
- Higher **holding cost $H$** decreases the optimal order quantity  


### Second Derivative

To verify that the critical point corresponds to a minimum of the total cost function, we compute the second derivative of $TC(Q)$ with respect to $Q$:

$$
\frac{d^2TC}{dQ^2} =
\frac{d}{dQ}\left(-\frac{DS}{Q^2} + \frac{H}{2}\right)
$$

Differentiating each term separately:

The derivative of the first term is:

$$
\frac{d}{dQ}\left(-\frac{DS}{Q^2}\right)
=
\frac{2DS}{Q^3}
$$

The derivative of the second term is:

$$
\frac{d}{dQ}\left(\frac{H}{2}\right)
=
0
$$

Therefore, the second derivative of the total cost function is:

$$
\frac{d^2TC}{dQ^2} =
\frac{2DS}{Q^3}
$$

Since \(D > 0\), \(S > 0\), and \(Q > 0\), it follows that:

$$
\frac{d^2TC}{dQ^2} > 0
$$

Because the second derivative is positive, the total cost function is convex at this point.  
Therefore, the critical point $Q^*$ corresponds to a minimum of the total cost function.

## 4. Model Assumptions and Constraints

## Assumptions of the Classical EOQ Model

The classical **Economic Order Quantity (EOQ)** model is based on several simplifying assumptions. These assumptions allow the model to be formulated and analyzed mathematically in a clear and consistent manner.

### Main Assumptions

#### Constant and Known Demand

The EOQ model assumes that the annual demand $D$ is known in advance and remains constant over time. This means that it does not change during the considered period and there are no fluctuations related to seasonality, market changes, or changes in consumer behavior. The demand flow is therefore uniform and predictable.

#### Fixed Ordering Cost

The ordering cost arises from administrative, logistical, and organizational activities related to processing an order. Therefore, the model assumes that each new order incurs the same fixed cost.

#### Constant Holding Cost

The holding cost per unit per year includes costs related to warehouse space, maintenance of storage facilities, the risk of damage or obsolescence of the goods, and costs associated with the deterioration of working capital. It is also assumed to remain constant.

#### No Stockouts

The model assumes that situations with insufficient or missing inventory do not occur. A new order arrives before the available inventory is completely depleted. The system always has enough products available to meet demand.

#### Instantaneous Replenishment

In the classical EOQ model it is assumed that once an order is placed, the goods arrive immediately or that the delivery time is negligible. This means that inventory is replenished instantly to the level of the new order, which simplifies the mathematical analysis of the model.

#### No Quantity Discounts

The model also assumes that the price per unit of product does not depend on the order quantity. In other words, situations in which the supplier offers discounts for larger orders are not considered. This allows the analysis to focus solely on the optimization of ordering and holding costs.

#### Single-Product Analysis

The classical EOQ model examines inventory management for a single product. Interactions between different products, limited storage space, or other factors that may arise in more complex inventory systems are not considered.

---

## Model Constraints

In order for the model to be mathematically and economically correct and meaningful, the parameters must satisfy the following conditions:

$$
D > 0, \quad S > 0, \quad H > 0, \quad Q > 0
$$

These constraints ensure that demand, ordering costs, and holding costs are positive values. In addition, the condition $Q > 0$ guarantees that the order quantity has real economic meaning. When these conditions are satisfied, the total cost function is well defined and can be used to determine the optimal order quantity.

## 5. Algorithm Design
## EOQ Algorithm Implementation

After the EOQ formula has been analytically derived through mathematical analysis, the next step is to transform the mathematical model into a simple algorithm.

---

### Inputs and Output

The algorithm uses the following inputs:

- $D$: annual demand  
- $S$: ordering cost per order  
- $H$: holding cost per unit per year  

The output is:

- $Q^*$: the Economic Order Quantity (EOQ)

---

### Computational Logic

The EOQ is calculated directly from the formula obtained during the analytical derivation of the model:

$$
Q^* = \sqrt{\frac{2DS}{H}}
$$

In addition to the optimal order quantity itself, it is often useful to define the total annual cost function:

$$
TC(Q) = \frac{D}{Q}S + \frac{Q}{2}H
$$

This function illustrates how total costs vary as a function of the order quantity $Q$. It enables numerical verification that the computed EOQ corresponds to the minimum total cost. For instance, the function's value can be evaluated for various values of $Q$ and the results compared.

---

### Input Validation

Before performing the calculations, it is necessary to ensure that the provided parameters are valid. For the model to be correct, they must be strictly positive.

Therefore, the following conditions must be satisfied:

- $D > 0$  
- $S > 0$  
- $H > 0$

Similarly, when calculating the total cost function for a given order quantity, the following condition must also be satisfied:

- $Q > 0$

If any of these parameters is equal to zero or has a negative value, the model loses its economic meaning and the calculations would not be correct. For this reason, input validation is an important part of any program implementation.

---

### Computational Complexity

The EOQ model is very efficient to compute. The optimal order quantity can be found with a simple calculation, without the need for loops or complex numerical methods.

Therefore, the computation has:

- **time complexity:** $O(1)$, since the result is obtained through a direct calculation  
- **space complexity:** $O(1)$, since no additional memory depending on the size of the input data is required  

Because of this, the EOQ model works well in real-world situations. It can be implemented in many programming environments for quick calculations and for analyzing inventory management policies.


## 6. Python Implementation

```python
def eoq(D, S, H):
    """
    Compute the EOQ.

    Parameters:
    D : float - Annual demand
    S : float - Ordering cost per order
    H : float - Holding cost per unit per year

    Returns:
    float - Optimal order quantity
    """

    if D <= 0 or S <= 0 or H <= 0:
        raise ValueError("D, S, and H must be positive numbers")

    return math.sqrt((2 * D * S) / H)
```

```python
def total_cost(Q, D, S, H):
    """
    Compute the TC.

    Parameters
    Q : float - Order quantity
    D : float - Annual demand
    S : float - Ordering cost
    H : float - Holding cost

    Returns
    float - Total annual cost
    """

    if Q <= 0:
        raise ValueError("Order quantity must be positive")

    ordering_cost = (D / Q) * S
    holding_cost = (Q / 2) * H

    return ordering_cost + holding_cost
```

```python
# Example parameters

D = 10000   # annual demand
S = 50      # ordering cost
H = 10       # holding cost
```

```python
Q_star = eoq(D, S, H)

print("EOQ:", round(Q_star, 2))
```

```python
cost_at_eoq = total_cost(Q_star, D, S, H)

print("TC at EOQ:", round(cost_at_eoq, 2))
```

## 7. Simulation and Cost Curve Visualization

```python
# Generate a range of order quantities

Q_values = np.linspace(100, 1500, 1000)

cost_values = [total_cost(Q, D, S, H) for Q in Q_values]
```

```python
plt.figure(figsize=(8,5))

plt.plot(Q_values, cost_values, label="Total Cost")

plt.axvline(Q_star, color="red", linestyle="--", label="EOQ")
plt.scatter(Q_star, cost_at_eoq, color="red")

plt.xlabel("Order Quantity (Q)")
plt.ylabel("Total Annual Cost")

plt.title("Total Cost vs Order Quantity")

plt.legend()
plt.grid(True)

plt.show()
```

The total cost curve has a characteristic U-shape.

A small order quantity results in frequent orders and higher ordering costs. Conversely, a large order quantity increases holding costs as inventory remains in storage longer.

The curve’s minimum point represents the Economic Order Quantity (EOQ), confirming that the earlier analytical solution minimizes total cost.

## 8. Numerical Verification Using Grid Search

In addition to the analytical derivation of EOQ, it is useful to verify the result numerically.

One effective method is to evaluate the total cost function across a range of order quantities and identify the quantity with the lowest cost.

This numerical approach does not use calculus. It directly tests multiple candidate values and compares the optimal result to the EOQ formula.

```python
# Numerical verification using grid search

Q_grid = np.arange(100, 1501, 1)
grid_costs = np.array([total_cost(Q, D, S, H) for Q in Q_grid])

Q_grid_min = Q_grid[np.argmin(grid_costs)]
TC_grid_min = grid_costs.min()

print("EOQ from formula:", round(Q_star, 2))
print("Minimum from grid search:", Q_grid_min)
print("Total cost at EOQ:", round(cost_at_eoq, 2))
print("Minimum total cost from grid search:", round(TC_grid_min, 2))
print("Difference between EOQ and grid minimum:", round(abs(Q_star - Q_grid_min), 2))
```

```python
plt.figure(figsize=(8,5))

plt.plot(Q_values, cost_values, label="Total Cost")
plt.axvline(Q_star, color="red", linestyle="--", label="EOQ")
plt.scatter(Q_grid_min, TC_grid_min, color="black", label="Grid Search Minimum")

plt.xlabel("Order Quantity (Q)")
plt.ylabel("Total Annual Cost")
plt.title("EOQ vs Numerical Minimum")
plt.legend()
plt.grid(True)

plt.show()
```

The grid search result closely matches the EOQ calculated using the analytical formula, confirming that the mathematically derived EOQ represents the minimum of the total cost function.

A small difference between the two values is expected because the grid search evaluates only discrete values of Q, while the analytical EOQ may be non-integer.
